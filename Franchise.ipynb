```python
import requests
import pandas as pd

# URL of the JSON data
url = 'https://ims-nv9i.onrender.com/franchises/all'

# Fetch data from the URL
response = requests.get(url)
data = response.json()

# Normalize and flatten JSON to a DataFrame
df = pd.json_normalize(data)

# Select key columns for analysis
df_selected = df[['FranchiseId', 'FranchiseName', 'City', 'Email', 'MobileNo', 'Approved', 'Active', 'Plan']]

# View the DataFrame
print(df_selected)

# Example: Analyze franchises by city
city_counts = df_selected['City'].value_counts()
print(city_counts)

# Example: Filter for approved franchises
approved_franchises = df_selected[df_selected['Approved'] == True]
print(approved_franchises)
```

```python
import requests
import pandas as pd

url = 'https://ims-nv9i.onrender.com/franchises/all'
response = requests.get(url)
data = response.json()

df = pd.json_normalize(data)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Option 2: Fill missing values (e.g., with a default value or mean)
df.fillna('Unknown')
```

```python
# Example: Analyze franchises by city
city_counts = df['City'].value_counts()
print(city_counts)

```

```python
import matplotlib.pyplot as plt

# Visualize franchise counts by city
city_counts.plot(kind='bar')
plt.title('Franchise Count by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

# Count franchises by plan type
plan_counts = df['Plan'].value_counts()
sns.barplot(x=plan_counts.index, y=plan_counts.values)
plt.title('Franchise Count by Plan')
plt.xlabel('Plan Type')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart of approved vs non-approved franchises
approved_counts = df['Approved'].value_counts()
approved_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Approved vs Non-Approved Franchises')
plt.ylabel('')
plt.show()
```

```python
# Bar chart of active vs inactive franchises
activity_counts = df['Active'].value_counts()
sns.barplot(x=activity_counts.index, y=activity_counts.values)
plt.title('Active vs Inactive Franchises')
plt.xlabel('Activity Status')
plt.ylabel('Count')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Plot number of franchises per year
franchises_per_year = df['Year'].value_counts().sort_index()
franchises_per_year.plot(kind='line', marker='o')
plt.title('Franchise Count Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Franchises')
plt.show()
```
